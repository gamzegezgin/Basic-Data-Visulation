```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('healthcare_dataset.csv')

#kategorik değişken: sütun grafik, countplot bar
#sayısal değişken: hist, boxplot
```

```python
df["Gender"].value_counts().plot(kind='bar')
plt.show()
```

```python
df["Blood Type"].value_counts().plot(kind='bar')
plt.xlabel('Blood Type')
plt.ylabel('Count of Patient')
plt.title('Number of Patients by Blood Type')
plt.ylim(6800, 7000)
plt.show()
```

```python
sns.catplot(x='Blood Type', hue = 'Gender', data=df, kind='count', palette='coolwarm', height=10, aspect=1)
plt.xlabel('Blood Type')
plt.ylabel('Count of Patient')
plt.title('Distribution of Blood Types by Gender')
plt.ylim(3000, 3600)
plt.show()
```

```python
plt.figure(figsize=(18, 8))
sns.catplot(x='Blood Type', hue = 'Medical Condition', data=df, kind='count', palette='coolwarm', height=10, aspect=1)
plt.xlabel('Blood Type')
plt.ylabel('Count of Patient')
plt.title('Distribution of Blood Types by Medical Condition')

plt.ylim(1100, 1250)

plt.show()
```

```python
#tarih sutunlarını tarih formatına cevirdik

df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
```

```python
#tarihlerin farkını alarak yatış süresini hesaplayacagız

df['Length of Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days

print(df[['Date of Admission', 'Discharge Date', 'Length of Stay']].head())
#bir koseli parantez dataframe için diğeri de liste yapısı için
```

```python
average_stay = df.groupby('Medical Condition')['Length of Stay'].mean().reset_index()

sns.barplot(x='Medical Condition', y='Length of Stay', data=average_stay)
plt.xlabel('Medical Condition')
plt.ylabel('Average Length of Stay (days)')
plt.title('Average Length of Stay by Medical Condition')
plt.ylim(14, 16)
plt.xticks(rotation=45)
plt.show()
```

```python
age_average = df.groupby('Medical Condition')['Age'].mean().reset_index(name='Average Age')

sns.barplot(x='Medical Condition', y='Average Age', data=age_average)
plt.xlabel('Medical Condition')
plt.ylabel('Average Age')
plt.title('Average Age by Medical Condition')
plt.xticks(rotation=45)
plt.show()
```
